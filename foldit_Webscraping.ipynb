```python
import requests
from bs4 import BeautifulSoup 
from csv import writer
```

```python
url = 'https://fold.it/leaderboards/Overall'
page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify)
```

```python
lists = soup.find_all('strong', class_ = 'h4 align-items-center d-flex')
print(lists)
```

```python
for list in lists:
    name =  list.find('a').text.replace('\n','')
    level = list.find('span', class_= 'badge badge-primary').text.replace('\n','')
    score = list.find('span', class_= 'ml-auto').text.replace('\n','')
    info = [name, score, level]
    print(info)
```

```python
with open('foldit_score_details.csv', 'a', encoding = 'UTF-8') as f:
    file = writer(f)
#     header = ['Name' , 'Score', 'Level']
#     file.writerow(header)
    for list in lists:
        name =  list.find('a').text.replace('\n','')
        level = list.find('span', class_= 'badge badge-primary').text.replace('\n','')
        score = list.find('span', class_= 'ml-auto').text.replace('\n','')
        info = [name, score, level]
        file.writerow(info)
    
```
